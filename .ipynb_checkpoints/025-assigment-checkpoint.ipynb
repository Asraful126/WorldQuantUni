```python
# Import libraries here

import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import VimeoVideo
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.utils.validation import check_is_fitted
import seaborn as sns
```

```python
import openpyxl
```

```python
df = pd.read_excel('Maxico_1.xlsx')
```

```python
df.head()
```

**1. Subset the data in the CSV file and return only apartments in Mexico City ("Distrito Federal") that cost less than $100,000.**

```python
# drop redundant col 'Unnamed: 0'
df = df.drop(columns='Unnamed: 0', axis='columns')
```

```python
# property type is apartment
df_prop = df['property_type'] == 'apartment'
```

```python
# place is Distrito Federal
df_place = df['place_with_parent_names'].str.contains("Distrito Federal")
```

```python
# usd approx price < 100000
df_price_usd = df['price_aprox_usd'] < 100_000
```

```python
df1 = df[df_prop & df_place & df_price_usd]
```

```python
df1.head()
```

**2. Remove outliers by trimming the bottom and top 10% of properties in terms of "surface_covered_in_m2".**

```python
# surface area between 10% and 90%
low, high = df1['surface_covered_in_m2'].quantile([0.1, 0.9])
df2 = df1[df1['surface_covered_in_m2'].between(low, high)]
```

```python
df2.head()
```

**3. Create separate "lat" and "lon" columns.**

```python
# split col [lat-lon] into [lat] & [lon]
df2[['lat', 'lon']] = df2['lat-lon'].str.split(',', expand=True).astype(float)
df3 = df2.drop(columns='lat-lon', axis='columns')
```

```python
df3.head()
```

**4. Mexico City is divided into 15 boroughs. Create a "borough" feature from the "place_with_parent_names" column.**

```python
# create df['borough'] from 'place_with_parent_names'
df3['borough'] = df3["place_with_parent_names"].str.split("|", expand=True)[1]
df4 = df3.drop(columns='place_with_parent_names', axis='columns')
```

```python
df4.head()
```

```python
df4.shape
```

**5. Drop columns that are more than 50% null values.**

```python
df4.isnull().sum()
```

```python
# Drop columns that are more than 50% null values.
df5 = df4.loc[:, df4.isnull().sum() <= df4.shape[0]/2]
```

```python
df5.head()
```

```python
df5.isnull().sum()
```

**6. Drop columns containing low- or high-cardinality categorical values.**

```python
df5.select_dtypes('object').head()
```

```python
df5.select_dtypes('object').nunique()
```

```python
# So operation, property_type, currency are of low cardinality
# property_url is of high cardinality
# both cardinality types do not give  a trend to be modeled
# drop them

df6 = df5.drop(columns=['operation',
                      'property_type',
                      'currency', 
                      'properati_url'], axis='columns')
```

```python
df6.head()
```

**7. Drop any columns that would constitute leakage for the target "price_aprox_usd".**

```python
 df7 = df6.drop(columns=['price', 'price_aprox_local_currency', 'price_per_m2'], axis='columns')
```

```python
df7.head()
```

**8. Drop any columns that would create issues of multicollinearity.**

```python
# Draw a heat map of corr except the target columns price_approx_usd
corr = df7.select_dtypes('number').drop(columns='price_aprox_usd').corr() 
sns.heatmap(corr)
```
